# Zalando scraper | Pre-owned heren

```python
## GOOD TO KNOW 

# website = https://zalando.nl/pre-owned-mode-heren/index.html
# article class = DT5BTM w8MdNG cYylcv QylWsg _75qWlu iOzucJ JT3_zV DvypSJ
# tag product brand names = <span> 
# inspecting a product gives us: <a class="_LM JT3_zV CKDt_l CKDt_l LyRfpJ"href="https://www.zalando.nl/g-star-overhemd-blue-zir01hbly-001.html" rel="">
```

```python
# Import packages
from bs4 import BeautifulSoup
import requests
import time 
from time import sleep
from datetime import datetime
import csv

# Make selenium and chromedriver work
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from re import search
from selenium.webdriver import ActionChains
```

```python
#debug = {'verbose': sys.stderr}
user_agent = {'User-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36 Edg/98.0.1108.62'}  
```

```python
# use webdriver.Chrome()
driver_po = webdriver.Chrome(ChromeDriverManager().install())

# run Zalando Pre-owned 
driver_po.get("https://www.zalando.nl/pre-owned-mode-heren/")
time.sleep(10)
print("done")
```

```python
# accept button on the website (close pop-ups before proceeding)
driver_po.find_element_by_id("uc-btn-accept-banner").click()
```

```python
# Create a function to collect all page urls
base_url_po = "https://www.zalando.nl/pre-owned-mode-heren/?p="
counter_po = 1

# Now create the function
def generate_page_urls_po(base_url_po, num_pages_po):
    page_urls_po = []
    
    for counter_po in range (1, num_pages_po + 1):
        full_url_po = base_url_po + str(counter_po) + "&order=activation_date"
        page_urls_po.append(full_url_po)
    
    return page_urls_po
```

```python
page_urls_po = generate_page_urls_po(base_url_po, 1) 

# Let's check and create 20 url's
generate_page_urls_po("https://www.zalando.nl/pre-owned-mode-heren/?p=", 20)
```

```python
# Create a function to collect the product_urls
def create_product_urls_po(page_urls_po):
    product_urls_po = []
    cnt=1
    for page_url_po in page_urls_po:
        cnt+=1
        driver_po.get(page_url_po)
        time.sleep(5)
        res_po = driver_po.page_source.encode('utf-8')
        soup_po = BeautifulSoup(res_po, "html.parser")
        products_po = soup_po.find_all("div", class_= "DT5BTM w8MdNG cYylcv QylWsg _75qWlu iOzucJ JT3_zV DvypSJ")       
        sleep(1)

        for product_po in products_po:
            try:
                product_url_po = product_po.find("a").attrs["href"]
                print(product_url_po)
                product_urls_po.append(product_url_po)
            except:
                print("dit is geen product")
                
    return product_urls_po
```

```python
product_urls_po = create_product_urls_po(page_urls_po)
```

```python
# scraping information from product pages

def get_product_pages_po(products_po):
    product_pages_po = []
    
    for url_po in product_urls_po: 
        res_po = requests.get(url_po, headers = user_agent)
        soup_po = BeautifulSoup(res_po.text, "html.parser")
        attributes_po = soup_po.find_all(class_="DT5BTM VHXqc_ rceRmQ _4NtqZU mIlIve")
        delivery_po = soup_po.find(class_="PBR5VH _0xLoFW EJ4MLB")
        prices_po = soup_po.find("div", class_="_0xLoFW vSgP6A _7ckuOK")
        
        for attribute_po in attributes_po:    
            product_po = attribute_po.find("h3", class_="mt1kvu ka2E9k uMhVZi FxZV-M SZKKsK pVrzNP _5Yd-hZ").text
            category_po = attribute_po.find("span", class_="EKabf7 R_QwOV").text
            color_po = attribute_po.find("span", class_="u-6V88 ka2E9k uMhVZi dgII7d RYghuO pVrzNP").text
            size_po = attribute_po.find("span", class_="_7Cm1F9 ka2E9k uMhVZi dgII7d RYghuO pVrzNP").text
            delivery_time_po = delivery_po.find("span", class_="u-6V88 ka2E9k uMhVZi dgII7d RYghuO pVrzNP").text
            delivery_time_po = delivery_time_po.replace("werkdagen", "")
            delivery_time_po = delivery_time_po.replace(" ", "")
            delivery_time_po = delivery_time_po.replace("-", "--")

            try:
                price_po = prices_po.find("span", class_="pVrzNP").text
                price_po = price_po.replace(u'€\xa0', u' ')
                price_po = price_po.replace(" ", "")
            except:
                price_po = prices_po.find("span", class_="uqkIZw").text
                price_po = price_po.replace(u'€\xa0', u' ')
                price_po = price_po.replace(" ", "")

            product_pages_po.append({"url": url_po,
                                 "brand": product_po,
                                 "category": category_po,
                                 "price": price_po,
                                 "size": size_po,
                                 "delivery time": delivery_time_po,
                                 "color": color_po})
   
    return product_pages_po
```

```python
product_attributes_po = get_product_pages_po(product_urls_po)

time.sleep(5)

product_attributes_po
```

```python
with open("product_description_pre_owned.csv", "w", encoding = 'utf-8', newline='') as csv_file: 
    writer = csv.writer(csv_file, delimiter = ";")
    writer.writerow(["url", "brand", "category", "price", "size", "delivery time", "color", "now"])
    now_po = datetime.now()
    for product_po in product_attributes_po: 
        writer.writerow([product_po['url'], product_po['brand'], product_po['category'], product_po['price'], product_po['size'], product_po['delivery time'], product_po["color"], now_po])
print('done!')
```

```python
# use webdriver.Chrome()
driver = webdriver.Chrome(ChromeDriverManager().install())

# run Zalando Herenkleding
driver.get("https://www.zalando.nl/herenkleding/")
time.sleep(10)
print("done")
```

```python
# Create a function to collect all page urls
base_url = "https://www.zalando.nl/herenkleding/?p="
counter = 1

# Now create the function
def generate_page_urls(base_url, num_pages):
    page_urls = []
    
    for counter in range (1, num_pages + 1):
        full_url = base_url + str(counter) + "&order=activation_date"
        page_urls.append(full_url)
    
    return page_urls
```

```python
page_urls = generate_page_urls(base_url, 1) 

generate_page_urls("https://www.zalando.nl/herenkleding/?p=", 20)
```

```python
# Create a function to collect the product_urls
def create_product_urls(page_urls):
    product_urls = []
    cnt=1
    for page_url in page_urls:
        cnt+=1
        driver.get(page_url)
        time.sleep(5)
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        products = soup.find_all("div", class_= "DT5BTM w8MdNG cYylcv QylWsg _75qWlu iOzucJ JT3_zV DvypSJ")       
        sleep(1)

        for product in products:
            try:
                product_url = product.find("a").attrs["href"]
                print(product_url)
                product_urls.append(product_url)
            except:
                print("dit is geen product")
                
    return product_urls
```

```python
product_urls = create_product_urls(page_urls)
```

```python
# scraping information from product pages - Herenkleding
def get_product_pages(products):
    product_pages = []
    
    for url in product_urls: 
        res = requests.get(url, headers = user_agent)
        soup = BeautifulSoup(res.text, "html.parser")
        attributes = soup.find_all(class_="DT5BTM VHXqc_ rceRmQ _4NtqZU mIlIve")
        delivery = soup.find(class_="PBR5VH _0xLoFW EJ4MLB")
        prices = soup.find("div", class_="_0xLoFW vSgP6A _7ckuOK")
        size = soup.find(class_="MU8FaS _0xLoFW VuYQuj parent _78xIQ-")
        
        for attribute in attributes:    
            product = attribute.find("h3", class_="mt1kvu ka2E9k uMhVZi FxZV-M SZKKsK pVrzNP _5Yd-hZ").text
            category = attribute.find("span", class_="EKabf7 R_QwOV").text
            color = attribute.find("span", class_="u-6V88 ka2E9k uMhVZi dgII7d RYghuO pVrzNP").text
            delivery_time = delivery.find("span", class_="u-6V88 ka2E9k uMhVZi dgII7d RYghuO pVrzNP").text
            delivery_time = delivery_time.replace("werkdagen", "")
            delivery_time = delivery_time.replace(" ", "")
            delivery_time = delivery_time.replace("-", "--")


            
            substring = " - "            
            if search(substring, category):
                category = category.rsplit(' -', 1)
                category = category.pop(1)
                # category = category.replace(" -", "")
                # regel die teskt ervoor verwijderd
            else:
                category
                
            try:
                price = prices.find("span", class_="pVrzNP").text
                price = price.replace(u'€\xa0', u' ')
                price = price.replace(" ", "")
            except:
                price = prices.find("span", class_="uqkIZw").text
                price = price.replace(u'€\xa0', u' ')
                price = price.replace(" ", "")
            
            product_pages.append({"url": url,
                                 "brand": product,
                                 "category": category,
                                 "price": price,
                                 "delivery time": delivery_time,
                                 "color": color})
   
    return product_pages
print("done")
```

```python
product_attributes = get_product_pages(product_urls)

time.sleep(5)

product_attributes
```

```python
# make csv
with open("product_description_herenkleding.csv", "w", encoding = 'utf-8', newline='') as csv_file: 
    writer = csv.writer(csv_file, delimiter = ";")
    writer.writerow(["url", "brand", "category", "price", "delivery time", "color", "now"])
    now = datetime.now()
    for product in product_attributes: 
        writer.writerow([product['url'], product['brand'], product['category'], product['price'], product['delivery time'], product['color'], now])
print('done!')
```
