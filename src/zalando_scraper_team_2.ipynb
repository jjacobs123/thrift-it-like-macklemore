# THRIFT IT LIKE MACKLEMORE @ ZALANDO

This webscraper contains two parts: the web scraper for Zalando's Pre-owned men's clothing section and the web scraper for Zalando's Regular men's clothing section. The purpose of these web scrapers is to gather information about the products and to determine the difference between the offerings of the Pre-owned and the Regular Zalando web pages based on brands, product categories and prices. The feasible sample size contains 214 pages for both men's clothing sections and 18190 products for Zalando's Pre-owned men's clothing section and 18190 products for Zalando's Regular men's clothing section.

**Import packages & run Selenium**  
Selenium is used to scrape the product information of the Zalando pages as the products on the web pages load when scrolling down the page. However, it is not possible to manually scroll down the pages when web scraping the website. 

```python
from bs4 import BeautifulSoup
import requests
import time 
from time import sleep
from datetime import datetime
import csv

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from re import search
from selenium.webdriver import ActionChains
```

```python
# debug = {'verbose': sys.stderr}
user_agent = {'User-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102 Safari/537.36 Edg/98.0.1108.62'}  
```

## Chapter 1: Zalando scraper | Pre-owned Heren
This part of the web scraper contains code to scrape the Zalando web page(s) of the Zalando Pre-owned Men's clothing section (https://zalando.nl/pre-owned-mode-heren/index.html). From now on, the variable names end with 'po' as an indication to the Pre-owned label.

### 1.1 Run Chromedriver

```python
# use webdriver.Chrome()
driver_po = webdriver.Chrome(ChromeDriverManager().install())

# run Zalando Pre-owned 
driver_po.get("https://www.zalando.nl/pre-owned-mode-heren/")
time.sleep(10)
print("done")
```

### 1.2 Automatic acceptation of cookies
The cookies are automatically accepted to close pop-ups before proceeding with web scraping.

```python
driver_po.find_element_by_id("uc-btn-accept-banner").click()
```

### 1.3 Collect page urls

```python
base_url_po = "https://www.zalando.nl/pre-owned-mode-heren/?p="
counter_po = 1

def generate_page_urls_po(base_url_po, num_pages_po):
    page_urls_po = []
    
    for counter_po in range (1, num_pages_po + 1):
        full_url_po = base_url_po + str(counter_po) + "&order=activation_date"
        page_urls_po.append(full_url_po)
    
    return page_urls_po
```

```python
page_urls_po = generate_page_urls_po(base_url_po, 214) 

# check and generate a subsample of 20 url's
generate_page_urls_po("https://www.zalando.nl/pre-owned-mode-heren/?p=", 20)
```

### 1.4 Collect product urls

```python
def create_product_urls_po(page_urls_po):
    product_urls_po = []
    cnt=1
    for page_url_po in page_urls_po:
        cnt+=1
        driver_po.get(page_url_po)
        time.sleep(5)
        res_po = driver_po.page_source.encode('utf-8')
        soup_po = BeautifulSoup(res_po, "html.parser")
        products_po = soup_po.find_all("div", class_= "DT5BTM w8MdNG cYylcv QylWsg _75qWlu iOzucJ JT3_zV DvypSJ") 
        print(len(products_po))
        sleep(1)

        for product_po in products_po:
            try:
                product_url_po = product_po.find("a").attrs["href"]
                print(product_url_po)
                product_urls_po.append(product_url_po)
            except:
                print("dit is geen product")
                
    return product_urls_po
```

```python
product_urls_po = create_product_urls_po(page_urls_po)
```

### 1.5 Collect information from products
The information gathered per product for the Zalando's Pre-owned men's clothing section includes product url, brand name, category type, color, size, delivery time in working days and price in euros. 

```python
def get_product_pages_po(products_po):
    product_pages_po = []
    
    for url_po in product_urls_po: 
        res_po = requests.get(url_po, headers = user_agent)
        soup_po = BeautifulSoup(res_po.text, "html.parser")
        attributes_po = soup_po.find_all(class_="DT5BTM VHXqc_ rceRmQ _4NtqZU mIlIve")
        delivery_po = soup_po.find(class_="PBR5VH _0xLoFW EJ4MLB")
        prices_po = soup_po.find("div", class_="_0xLoFW vSgP6A _7ckuOK")
        
        for attribute_po in attributes_po:    
            product_po = attribute_po.find("h3", class_="SZKKsK mt1kvu ka2E9k uMhVZi FxZV-M pVrzNP _5Yd-hZ").text
            category_po = attribute_po.find("span", class_="EKabf7 R_QwOV").text
            color_po = attribute_po.find("span", class_="RYghuO u-6V88 ka2E9k uMhVZi dgII7d pVrzNP").text
            size_po = attribute_po.find("span", class_="RYghuO _7Cm1F9 ka2E9k uMhVZi dgII7d pVrzNP").text
            
            try:
                delivery_time_po = delivery_po.find("span", class_="u-6V88 ka2E9k uMhVZi dgII7d RYghuO pVrzNP").text
                delivery_time_po = delivery_time_po.replace("werkdagen", "")
                delivery_time_po = delivery_time_po.replace(" ", "")
                delivery_time_po = delivery_time_po.replace("-", "--")
            except:
                delivery_time_po = "no delivery time"

            try:
                price_po = prices_po.find("span", class_="pVrzNP").text
                price_po = price_po.replace(u'€\xa0', u' ')
                price_po = price_po.replace(" ", "")
                price_po = price_po.replace(",", ".")
            except:
                price_po = prices_po.find("span", class_="uqkIZw").text
                price_po = price_po.replace(u'€\xa0', u' ')
                price_po = price_po.replace(" ", "")
                price_po = price_po.replace(",", ".")

            product_pages_po.append({"url": url_po,
                                 "brand": product_po,
                                 "category": category_po,
                                 "price": price_po,
                                 "size": size_po,
                                 "delivery_time": delivery_time_po,
                                 "color": color_po})
   
    return product_pages_po
```

```python
product_attributes_po = get_product_pages_po(product_urls_po)

time.sleep(5)

product_attributes_po
```

### 1.6 Write collected information to a CSV file

```python
with open("product_description_pre_owned.csv", "w", encoding = 'utf-8', newline='') as csv_file: 
    writer = csv.writer(csv_file, delimiter = ";")
    writer.writerow(["url", "brand", "category", "price", "size", "delivery_time", "color", "date"])
    now_po = datetime.now()
    for product_po in product_attributes_po: 
        writer.writerow([product_po['url'], product_po['brand'], product_po['category'], product_po['price'], product_po['size'], product_po['delivery_time'], product_po["color"], now_po])
print('done!')
```

##  Chapter 2: Zalando scraper | Regular Heren
The second part of the web scraper contains code to scrape the Zalando web page(s) of the Zalando Regular Men's clothing section (https://www.zalando.nl/herenkleding/). The variables used as an outcome of this part of the scraper do not have a particular denote as indication.

### 2.1 Run Chromedriver

```python
# use webdriver.Chrome()
driver = webdriver.Chrome(ChromeDriverManager().install())

# run Zalando Herenkleding
driver.get("https://www.zalando.nl/herenkleding/")
time.sleep(10)
print("done")
```

### 2.2 Automatic acceptation of cookies
The cookies are automatically accepted to close pop-ups before proceeding with web scraping.

```python
driver.find_element_by_id("uc-btn-accept-banner").click()
```

### 2.3 Collect page urls

```python
base_url = "https://www.zalando.nl/herenkleding/?p="
counter = 1

def generate_page_urls(base_url, num_pages):
    page_urls = []
    
    for counter in range (1, num_pages + 1):
        full_url = base_url + str(counter) + "&order=activation_date"
        page_urls.append(full_url)
    
    return page_urls
```

```python
page_urls = generate_page_urls(base_url, 214) 

generate_page_urls("https://www.zalando.nl/herenkleding/?p=", 20)
```

### 2.4 Collect product urls

```python
def create_product_urls(page_urls):
    product_urls = []
    cnt=1
    for page_url in page_urls:
        cnt+=1
        driver.get(page_url)
        time.sleep(5)
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        products = soup.find_all("div", class_= "DT5BTM w8MdNG cYylcv QylWsg _75qWlu iOzucJ JT3_zV DvypSJ")
        substring_outfit = "/outfits/"
        print(len(products))
        sleep(1)

        for product in products:
            try:
                product_url = product.find("a").attrs["href"]
                
                if search(substring_outfit, product_url):
                    product_url = product_url.rsplit('/outfits', 1)
                    product_url = product_url.pop(0)
                else:
                    print(product_url)
                    product_urls.append(product_url)

            except:
                print("dit is geen product")
                
    return product_urls
```

```python
product_urls = create_product_urls(page_urls)
```

### 2.5 Collect information from products
The information gathered per product for the Zalando's Regular men's clothing section includes product url, brand name, category type, color, delivery time in working days and price in euros.   
*NOTE: Since collecting the sizes was made complicated by a dropdown menu, it was not possible to scrape this information. The assumptions was made that for the regular offering of men's clothing all sizes are available.*

```python
def get_product_pages(products):
    product_pages = []
    
    for url in product_urls: 
        res = requests.get(url, headers = user_agent)
        soup = BeautifulSoup(res.text, "html.parser")
        attributes = soup.find_all(class_="DT5BTM VHXqc_ rceRmQ _4NtqZU mIlIve")
        delivery = soup.find(class_="PBR5VH _0xLoFW EJ4MLB")
        prices = soup.find("div", class_="_0xLoFW vSgP6A _7ckuOK")
        
        for attribute in attributes:    
            product = attribute.find("h3", class_="SZKKsK mt1kvu ka2E9k uMhVZi FxZV-M pVrzNP _5Yd-hZ").text
            category = attribute.find("span", class_="EKabf7 R_QwOV").text
            color = attribute.find("span", class_="RYghuO u-6V88 ka2E9k uMhVZi dgII7d pVrzNP").text
            
            try:
                delivery_time = delivery.find("span", class_="RYghuO u-6V88 ka2E9k uMhVZi dgII7d pVrzNP").text
                delivery_time = delivery_time.replace("werkdagen", "")
                delivery_time = delivery_time.replace(" ", "")
                delivery_time = delivery_time.replace("-", "--")
            except:
                delivery_time = "no delivery time"
            
            substring = " - "            
            if search(substring, category):
                category = category.rsplit(' -', 1)
                category = category.pop(1)
                
            else:
                category
                
            try:
                price = prices.find("span", class_="pVrzNP").text
                price = price.replace(u'€\xa0', u' ')
                price = price.replace(" ", "")
            except:
                price = prices.find("span", class_="uqkIZw").text
                price = price.replace(u'€\xa0', u' ')
                price = price.replace(" ", "")
            
            product_pages.append({"url": url,
                                 "brand": product,
                                 "category": category,
                                 "price": price,
                                 "delivery_time": delivery_time,
                                 "color": color})
   
    return product_pages
print("done")
```

### 2.6 Write collected information to a CSV file

```python
with open("product_description_herenkleding.csv", "w", encoding = 'utf-8', newline='') as csv_file: 
    writer = csv.writer(csv_file, delimiter = ";")
    writer.writerow(["url", "brand", "category", "price", "delivery_time", "color", "date"])
    now = datetime.now()
    for product in product_attributes: 
        writer.writerow([product['url'], product['brand'], product['category'], product['price'], product['delivery_time'], product['color'], now])
print('done!')
```
